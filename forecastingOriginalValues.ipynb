```python
import pandas as pd

df = pd.read_csv("/content/processed_df.csv")
display(df.head())
display(df.info())
```

```python
# Replace the file path prefix
df['processed_image_path'] = df['processed_image_path'].str.replace('/content/drive/MyDrive/Colab Notebooks/Unilag project/fapar_patches_processed/', 'fapar_patches_processed/')

# Drop rows where 'hs_name' is "Not assessed"
df = df[df['hs_name'] != "Not assessed"].copy()

# Set 'hs_code' to 1 for rows where 'hs_name' is "Major hotspot"
df.loc[df['hs_name'] == "Major hotspot", 'hs_code'] = 1

display(df.head())
display(df.info())
```

```python
df.drop(columns=['iso3', 'year', 'hs_code', 'hs_name', 'processed_image_path']).columns
```

```python
train_df = df.sample(frac=0.8, random_state=42)
test_df = df.drop(train_df.index)

train_files = train_df['processed_image_path'].values
train_labels = train_df['hs_code'].values

test_files = test_df['processed_image_path'].values
test_labels = test_df['hs_code'].values
```

```python
from sklearn.model_selection import train_test_split

train_files, test_files, train_features, test_features, train_labels, test_labels = train_test_split(df['processed_image_path'].values, df.drop(columns=['iso3', 'year', 'hs_code', 'hs_name', 'processed_image_path']).values, df['hs_code'].values, test_size=0.2, random_state=42, stratify=df['hs_code'].values)
```

```python
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import tifffile as tiff
train_images = []
for file in tqdm(train_files, total = len(train_files)):
  train_images.append(cv2.resize(tiff.imread(file), ( 64, 64)))
train_images = np.array(train_images)

test_images = []
for file in tqdm(test_files, total = len(test_files)):
  test_images.append(cv2.resize(tiff.imread(file), ( 64, 64)))
test_images = np.array(test_images)
```
